```python
# Extract the x and y data
x = df['Cocoa Exports in Tons']
y = df['Population']

# Scatter plot of 'Cocoa Exports in Tons' vs 'Population'
plt.figure(figsize=(8,6))
plt.scatter(x, y, color='blue', label='Data points')

# Calculate the line of best fit
slope, intercept = np.polyfit(x, y, 1)
line = slope * x + intercept

# Plot the best-fit line
plt.plot(x, line, color='red', label=f'Best-fit line: y={slope:.2f}x + {intercept:.2f}')

# Add titles and labels
plt.title('Cocoa Exports in Tons vs Population with Correlation Line')
plt.xlabel('Cocoa Exports in Tons')
plt.ylabel('Population')

# Add a legend
plt.legend()

# Display the grid and the plot
plt.grid(True)
plt.show()
```
